# 문제 3

```python
data = pd.read_csv('data/Traffic_Accident_2017.csv', encoding='euc-kr')
```

```python
data['사고유형_대분류'].value_counts()
```

```python
data['주야'].value_counts()
```

```python
data['요일'].value_counts()
```

```python
ac = data[data['사고유형_대분류']== '차대사람']
ac
```

```python
ac3 = ac[ac['주야']== '야간']
ac3
```

```python
days = ac3['요일'].value_counts()
days
```

```python
days = days.reindex(['월','화','수','목','금','토','일'])
days
```

```python
days.index
```

```python
# x, y축 값 설정하기
x = np.arange(7)



#그래프 그리기
plt.bar(x, days)
plt.xticks(x, labels =days.index)
plt.xlabel("요일")
plt.ylabel("사고 건수")
plt.title("요일별 야간 차대사람 교통사고 건수")
plt.ylim(0,200)

plt.show()
```
