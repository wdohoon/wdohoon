#for 문
- 문자열 또는 리스트 또는 튜플이 들어갔을 때 안에 있는 요소를 하나씩 반복
- for 변수 in 문자열(or 리스트 or 튜플) :
  print(변수)

```python
hi = "안녕하세요"
for s in hi :
    print(s)
```

```python
tuple_food = ("햄버거","치킨","피자")
for food in tuple_food:
    print(food)
```

```python
score_list = [90,45,70,60,55]
num = 0
for score in score_list:
    num += 1
    if score >= 60 :
        print(f"{num}번 학생은 합격입니다.")
    else :
        print(f"{num}번 학생은 불합격입니다")
```

```python
python_score = [57,86,63,92,35,79]
num = 0
for socre in python_score :
    num += 1
    if socre >= 80 :
        print(f"{num}번은 {socre}점 이며, A등급 입니다.")
    elif socre >= 60 :
        print(f"{num}번은 {socre}점 이며, B등급 입니다.")
    else :
        print(f"{num}번은 {socre}점 이며, C등급 입니다.")
```

```python
for socre in python_score : 
    num += 1
    if socre >= 80:
        grade = 'A'
    elif socre >= 60:
        grade = 'B'
    else :
        grade = 'C'
        
    print(f"{num}번은 {socre}점 이며, {grade}등급 입니다.")
```

# rage()함수 사용
- 필요한 만큼의 숫자를 만들어내는 유용한 기능
- range(시작할 숫자,종료할 숫자,증감량)
- range(1,10,1) -> 1부터 9까지 1씩 증가
- range(1,100,3) -> 1부터 99까지 3씩 증가
- range(10,1,-1) -> 10부터 2까지 1씩 감소(-1씩 증가)

```python
for i in range(1,10,1):
    print(i)
```

```python
num = 0
number1 = int(input("첫 번째 정수 입력>>"))
number2 = int(input("두 번째 정수 입력>>"))
number3 = int(input("배수 입력>>"))

for i in range(number1 , number2 + 1):
    if i % number3 == 0:
        num+=i
print(f"{number3}의 배수의 합{num}")
```

```python
quiz = input("OX 입력 >> ")
score = 0
count = 0

for result in quiz:
    if result == '0':
        count += 1
        score += count
    else :
        count = 0
print(score)
```

```python
ox = input("OX입력 >>")
score, total = 0, 0

for i in range(len(ox)):
    if ox[i] == 'O':
        score += 1
        total += score
    else : 
        score = 0
print(f"{score}점")
```
