```python
#자판기 실습

print("="*36)
print("="*10,"-"*3,"자판기","-"*3,"="*10)
money = int(input("insert coin >> "))
menu = int(input("메뉴 선택 >> "))

# 지불한 금액이 600원 이상일 경우
if money >= 600 :
    #1번 메뉴 선택
    if menu == 1 :
        money -= 600    #money = money - 600
    # 2번 메뉴 선택
    elif menu == 2 :
        # 넣은 돈이 800원 이상일 경우
        if money >= 800 :
            money -= 800
        # 넣은 돈이 600이상 800미만일 경우
        else : 
            print ("돈 더 넣으세요.")
    # 3번 메뉴 선택
    elif menu == 3 :
        # 넣은 돈이 1000원 이상일 경우
        if money >= 1000 :
            money -= 1000
        # 넣은 돈이 600이상 1000미만일 경우
        else : 
            print ("돈 더 넣으세요.")
    #1,2,3번이 아닌 다른 메뉴를 선택 했을 경우
    else :
        print("잘못된 메뉴")
        
        
    #잔돈 처리 (1000원,500원,100원으로 )
    cash_1000 = money // 1000
    cash_500 = money % 1000 // 500
    cash_100 = money % 1000 % 500 // 100
    
    print(f"잔돈 >> 1000원 {cash_1000}개, 500원 {cash_500}개, 100원{cash_100}개")
    
# 넣은돈이 600원 미만일 경우
else :
    print("돈을 더 넣어주세요")

    
```

- 추가,수정,삭제 불가능
- 리스트에 비해 동작 속도가 빠름

```python
tuple1 = (0, 1, 2, 3, ('a','b','c'),5)
tuple1
```

# 리스트와 튜플의 공통점과 차이점
#공통점
- 타입과 상관 없이 일련의 요소들을 가짐
- 요소의 순서를 관리
#차이점
- 리스트는 가변적이며 튜플은 불변적
- 리스트는 요소의 개수가 명확하지 않고 유동적일 경우에 사용
- 튜플은 요소 개수를 사전에 정확히 알고 있고 변동이 없을 경우에 사용(리스트에 비해 시스템 상 소모되는 메모리가 적고 속도가 빠름)

# 반복문

- 프로그램 내에서 똑같은 명령을 일정 횟수만큼 반복하여 수행하도록 제어하는 명령문

- 반복문 종류는 while문,for문이 있다.
- while : 반복 횟수가 명확하지 않을 때
- for : 반복 횟수가 명확할 때

```python
number = 1
while number <= 3:
    print(number)
    number += 1
```

```python
num = 0
while num < 10 :
    print("파이썬 최고!!")
    num += 1
```

```python
while True :
    print("무한루프")
    break
```

```python
number = 1
while True :
    print(number)
    number += 1
    if number > 3:
        break
```

```python
#continue - 반복문 시작으로 돌아가는 명령
num = 1
while num <= 10 :
    print(f"{num}번째 반복중")
    if num >= 4 :
        print("여기서부터 num 반복되지않음.")
        continue
    num += 1
    break
```

```python
while True :
    name = input("이름 입력 >>")
    if name == "최성우" :
        print("최성우 멋지다!")
        break
    else : 
        print(f"{name} 최고!")
```

```python
while True :
    num1 = int(input("첫 번째 정수 입력 >>"))
    num2 = int(input("두 번째 정수 입력 >>"))
    num3 = num1 + num2
    
    if num3 == 0 :
        print("프로그램이 종료되었습니다.")
        break
    else :
        print(f"두 정수의 합 : {num3}")
```

```python
while True :
    num1 = int(input("첫 번째 정수 입력 >>"))
    num2 = int(input("두 번째 정수 입력 >>"))
    
    if num1 ==0 and num2 ==0 :
        print("프로그램이 종료되었습니다.")
        break
        
    print(f"두 정수의 합 : {num1+num2}")
```

```python
#다이어트 관리 프로그램
weight = int(input("현재 몸무게 :"))
goal = int(input("목표 몸무게 :"))
week = 1

while weight >= goal :
    weight_week = int(input(f"{week}주차 감량 몸무게 : "))
    week += 1
    weight -= weight_week      # weight = weight - weight_week
print(f"{weight}kg 달성 축하합니다!")
```

```python
import random
num = random.randint(1,50)

while True :
    num1 = int(input("숫자를 입력하세요 >>"))
    
    if num > num1 :
        print(f"{num1}보다 큰 수 입니다.")
    elif num < num1 :
        print(f"{num1}보다 작은 수 입니다.")
    else :
        print("정답을 맞추셨습니다.")
        break
```
