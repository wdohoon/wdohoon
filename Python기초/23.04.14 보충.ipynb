#조건문 (if/elif/else)

```python
# 사전적 의미
# 프로그램에서 특정 조건이 참인지 혹은 거짓인지를 검사
# 코드의 실행 흐름을 제어하는 기능을 가진 구조

# 조건문은 분기점이다.
# 공통점 -> 특정 조건에 따라 다른 길을 선택
```

```python
# 프로그램을 실행 (컴퓨터가 판단)
    # 조건이 참이면 -> 특정코드를 실행
    # 조건이 거짓이면 -> 다른 코드를 실행 
```

```python
if True :
    print("실행문장 : 집가자~~")
if False :
    print("실행문장 : 언제가 ㅠㅠ")
print("if문 밖에 있는 실행문장 : '아모르직다'")
```

# Practice - 01

x = 5 로 선언해주고, 10이랑 비교하는 조건문을 써주세요.
그리고 아래와 같이 출력해주세요:

1. 10보다 클 경우: 'x는 10보다 큽니다'
2. 10과 같을 경우: 'x는 10과 같습니다'
3. 10보다 작을 경우 'x는 10보다 작습니다'

```python
x = 5 

if x > 10:
    print("x는 10보다 큽니다")
elif x == 10:
    print("x는 10과 같습니다")
else :
    print("x는 10보다 작습니다")
```

하지만 마음이 약해져서, 그래도 고소공포증이 있는 사람은 예약이 불가능하게 구축하려고 한다.

이름, 나이, 현대카드 소지 여부, 고소공포증 여부를 입력 받아 번지점프 코스를  
예약하는프로그램을 만들어 봅시다. 

요구사항 정의:

- 이름
- 나이
- 현대카드 소지 여부
- 고소공포증 여부

성인 기본 요금은 100,000만원, 미성년의 요금은 10%
성인은 현대카드 결재 시, 할인율은 10% (미성년자 이중할인 불가능) 

출력: 김윤호님의 총 이용료는 100000원입니다.

```python
# input을 문자열로:

name = input("이름 입력>>")
age = int(input("나이 입력>> "))
hyundai_card = input("현대카드 소지여부 (y/n)>> ")
phobia = input("고소공포증여부 (y/n)>> ")
fee = 100000


# 미성년자의 경우
if age < 20:
    fee *= 0.9 
    # fee = fee * 0.9

# 현대카드 소지 유무
elif hyundai_card == 'y':
     fee = fee * 0.9

# 고소공포증의 유무
if phobia == 'y':
    print('고소공포증이 있으시면 안전상 예약이 불가능합니다.')

else:
    print(f"{name}님의 총 이용료는 {int(fee)}원입니다. ")
```

?

결과는 이렇게 나와야한다:

hint) for문 -> if / else

```python
list1 = [1,30,7,9,17,3,5,23,44]

s = 0
d = 0

for i in list1 :
    if i% 2 == 1 :
        s += 1
    else :
        d += 1
print(f"짝수{d}개, 홀수{s}개")
```

# Practice - 04 

윤호쌤은 지금까지 열공한 연수생 중 4 명에게 
파격적인 커피 + 배달 서비스를 제안하고자 한다.

가장 빨리 답을 맞춘 이에게 보상이 주어진다.

우선 게임을 만들어보자:

랜덤으로 1부터 70사이의 숫자를 뽑으면, 뽑은 숫자를 맞추는 게임!
숫자를 맞추면 “당첨!” 를 출력하고 입력을 멈추세요!

```python

import random

#range(1,70)
#randint(1,71)

#range와 함수와 randit함수는 그 특성이 다릅니다.
#range함수는 종료값 포함x, but randit 함수는 시작값과 종료값을 모두 포함합니다. (1,70)

random_num = random.randint(1,70)

while True :
    # 점수 입력 받음
    input_num = int(input("점수 입력 : "))
    
    if random_num > input_num :
        print(f"{input_num}보다 큰 수 입니다.")
    elif random_num < input_num :
        print(f"{input_num}보다 작은 수 입니다.")
    else :
        print("정답")
        break
```
