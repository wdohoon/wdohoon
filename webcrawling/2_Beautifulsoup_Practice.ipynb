# 파싱 (Parsing)이란?
- 웹페이지에서 원하는 데이터를 추출, 가공하기 쉬운 상태로 변환하는 것
- ex) 생선을 낚았다, 최종적으로 통조림의 형태로 만들어야하는데, 손질하려는 절차

생선과는 달리, 웹페이지에 존재하는 데이터는 list.dictionary와 같은 자료구조와 다르다

그래서 parser라는 함수나 프로그램으로 다루기 쉬운 형태로 바꾸어 주는데

이 과정을 '파싱'이라고 한다

```python
# res_google
    # 그 응답받은 데이터를 , beautifulsoup을 사용해서 변환을 해볼겁니다.
    
bs_google = bs(res_google.text,'lxml')
bs_google
```

```python
# 자료 유형 알아보기
print(type(res_google))
print(type(bs_google))

# beautifulsoup으로 객체화가 진행이 됨,
```

## Pratice 1
- Google Tab에서 'Google' text 데이터를 가져오자

```python
soup = bs_google.select_one('title')
# bs_google에서 select_one 함수를 써서 하나를 선택할 건데,
    # 'title' tag selector로 뽑아주고 
        # soup이란 변수에 담아줘
soup
```

```python
soup.text
```
