```python
# CSS selector로 차량 명'들' 추출
suv_names = bs_cadi.select('h4.q-headline.none-margin')
suv_names[1].text
```

```python
# \n, \t => 이스케이프 문자 [Escape Character]
    # \n - 새로운 줄(exnter)
    # \t - 줄바꿈(tab)
# '\n\t\t\tSPORT PLATINUM\n\t\t'의 경우
# 줄바꿈 3개의 탭, 줄바꿈 2개의 탭이 있음을 코드로 나타냄
```

```python
# replace()함수를 사용, strip()함수를 사용할 겁니다.

# for loop

for name in suv_names:
    suv_sports = name.text.replace('\n','').strip()
    print(suv_sports)

# replace만 하면 끝?
    # 공백을 제거한 것처럼 보이나, 실제로 공백이 완전히 제거되지 않았음
    # 그래서 replace만 진행이 된다면, 개행문자가 빈 문자로 대체될 뿐,
        # 아직 공백은 사라지게 되는 것이 아님,
# 그래서 for loop 돌리는 김에, strip()함수까지 써서 한큐에 보내야함
```
