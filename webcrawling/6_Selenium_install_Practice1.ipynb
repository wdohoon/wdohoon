# Library Import

```python
# 컴퓨터용 브라우저 할당
from selenium import webdriver as wb

# 컴퓨터용 키보드 할당 - Keys의 'K'는 대문자
from selenium.webdriver.common.keys import Keys
    # 컴퓨터에겐 물리적인 키보드가 주어지지 않는다.
    # 그래서 우리는 컴퓨터에게 키 입력이 가능한 키보드를 선사한다.

# 구분자 역할을 해줌 (B는 대문자)
from selenium.webdriver.common.by import By
```

# 네이버에서 검색어 입력 후, Enter까지 해보자

```python
# 사용자가 크롬창을 열듯 
    # 브라우저에게 크롬창을 열어달라고 하기 -> 컴퓨터가 Chrome 브라우저를 제어
driver = wb.Chrome()
```

```python
# Requests 라이브러리에서 get을 통해 불러왔었다.
    # User-Agent가 필요하지 않은가?
        # 셀레니움에서 브라우저 - 브라우저로 요청하고 있기에 불필요.
```

```python
driver.get('https://www.naver.com')
```

### 검색창으로 접근

```python
# query

# find_element / find_elements
    # 헷갈린다? -> BeautifulSoup에 select와 select_one의 개념을 한번 생각해보자

    # Xpath 절대 경로
    # 최소 어디서 긁어왔는지 그 경로를 주석처리해주자
    # 편리하긴하지만, 항상 완벽하진 않습니다.
        # 업데이트가 잦은 요소인지 다시한번 확인해보고 사용하자.
        
search = driver.find_element(By.XPATH,'//*[@id="query"]') 
```

### 검색어를 입력 후 ENTER키 누르기

```python
# 검색어를 입력해보자
search.send_keys('오늘의운세')
```

```python
# 컴퓨터는 센스가 없다..
    # 검색어를 입력햇다고 해서, 센스있게 검색버튼, ENTER까지 친절하게 X
# 하지만 컴퓨터는 키보드, 마우스가 없다는 것을 기억하자 
# 그래서 코드로 명령을 하달해야한다.

search.send_keys(Keys.ENTER)

# 항상 자동제어창이 살아있는 것을 확인하자
```

### 뒤로가기

```python
# 검색을 하고, 다시 메인페이지로 가고싶다.
# 코드로 작성을 해줘야 합니다.

driver.back()

# 자동 브라우저에서 이전 페이지로 이동하는 method
# '뒤로가기'
    # HTML 코드가 아닌, 브라우저에서 제공하는 편의기능
    # driver로 선언한 자동제어창에게 코드로 부탁을 해야합니다.
```

### 재검색을 원할 경우

```python
# 검색어 창 지정부터, 검색어 입력, 검색 버튼 클릭까지 
#다시 지정을 해줘야 한다.

# 1. 다시 한번 검색창을 지정
search = driver.find_element(By.XPATH,'//*[@id="query"]') 
# 2. 다시 검색어 입력
search.send_keys('오늘의운세')
# 3. 다시 엔터
search.send_keys(Keys.ENTER)
# 4. 만약 다시 뒤로가기를 원한다면,
driver.back()
```

# NAVER에서 검색버튼을 클릭해보자

```python
# 크롬 브라우저 열기 및 네이버로 이동
driver = wb.Chrome()
driver.get('https://www.naver.com')
```

```python
# 검색창 요소 선택 - Xpath
    # 현재 사용자가 마우스를 검색창에 올려놓은 상태
search = driver.find_element(By.XPATH,'//*[@id="query"]')

#  검색어를 입력
search.send_keys('꽃가루 지수')

# 검색버튼을 Xpath로 추출해보자

btn = driver.find_element(By.XPATH,'//*[@id="search_btn"]')

# 컴퓨터는 마우스가 없다
    # 그래서 클릭을 하도록 코드로 명령을 내려줘야 한다
btn.click()
```

# 끝까지 스크롤해보기

```python
# 우선 body 태그를 body 변수에 지정을 하자

body = driver.find_element(By.TAG_NAME, 'body')

# 왜 body를 지정해가면서 해야할까?
    # 앞으로 크롤링 할 수 있는 데이터들이 대부분 body tag 안에 들어가있음
    # 스크롤링을 수행하기 위해서는, <body>태그가 선택되어야 한다.

body.send_keys(Keys.END)
```

### 크롬브라우저 종료

```python
# 사용자는 크롬에서 볼일 다 봤으면,
    # 크롬창을 닫습니다. -> '닫기버튼'
    # 닫기버튼은 브라우저에서 제공하는 기능
    # Selenium에서 이를 이용하여 해당 요소를 조작하기 위해서는, 다른 방법 사용

# 첫번째
#drive.close()
    # 현재 열려있는 창을 닫습니다.
# 두번째
driver.quit()
    # 현재 열려있는 창을 닫되,
    # 모든 webDriver 인스턴스와 관련된 모든 창,탭,프로세스 그리고 메모리를 종료
```

# 개인 실습 - 01
1. 크롬 브라우저를 켜고
2. 네이버로 접속
3. 검색창을 CSS_SELECTOR로 잡기 - 원하는 검색어 입력 (ENTER 키가 아닌 돋보기 아이콘 클릭)
4. 스크롤 끝까지 내리기
5. 뒤로 가기 버튼을 눌러서 다시 메인으로 이동
6. 크롬창 종료 - 메모리 누수까지 신경쓸 것!

```python
# 크롬 브라우저 켜기
driver = wb.Chrome()
# 네이버로 접속
driver.get('https://www.naver.com')
# 검색창을 CSS_SELECTOR로 잡기
search = driver.find_element(By.CSS_SELECTOR,'#query')
search.send_keys('adsp')
btn = driver.find_element(By.XPATH,'//*[@id="search_btn"]')
btn.click()
# 스크롤 끝까지 내리기
body = driver.find_element(By.TAG_NAME, 'body')
body.send_keys(Keys.END)
# 뒤로가기 버튼을 눌러서 다시 메인으로 이동
driver.back()
# 크롬창 종료 
driver.quit()
```

# 개인 실습 - 02
1. 크롬 브라우저를 켜고
2. 네이버로 접속
3. 검색창을 CSS_SELECTOR로 잡기 - 원하는 검색어 입력(ENTER 키 사용)
4. 검색 후 '이미지 탭' 클릭(CSS_SELECTOR,혹은 다른 방법도 사용가능)
5. 뒤로가기
6. 메인에서'뉴스 탭'클릭해서 들어가기 (CSS_SELECTOR,혹은 다른 방법도 사용 가능)
7. 뒤로 가기 버튼을 눌러서 다시 메인으로 이동
8. 크롬창 종료 - 메모리 누수까지 신경쓸 것!

```python
# 크롬 브라우저 켜기'
driver = wb.Chrome()
# 네이버 접속'
driver.get('https://www.naver.com')
# 검색창을 CSS_SELECTOR로 잡기
search = driver.find_element(By.CSS_SELECTOR,'#query')
search.send_keys('카누')
search.send_keys(Keys.ENTER)
#btn = driver.find_element(By.CSS_SELECTOR,'#search_btn')
#btn.click()
# 검색 후 '이미지 탭' 클릭
kanu = driver.find_element(By.CSS_SELECTOR,'#lnb > div.lnb_group > div > ul > li:nth-child(3) > a')
kanu.click()
# 뒤로가기
driver.back()
# 메인에서 '뉴스 탭' 클릭
kanu = driver.find_element(By.CSS_SELECTOR,'#lnb > div.lnb_group > div > ul > li:nth-child(8) > a')
kanu.click()
# 뒤로가기
driver.back()
# 크롬창 종료
driver.quit()
```
