```python
# 라이브러리 import
from selenium import webdriver as wb
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

import time
import pandas as pd
from tqdm import tqdm
```

### 브라우저 오픈

```python
driver = wb.Chrome()
driver.get('https://www.hsd.co.kr/')

# 햄버거 버튼 발생
    # 축소되면서 메뉴 버튼이 바뀌었는데,
    # 창을 최대화해서 진행
    
driver.maximize_window()
```

```python
# menu -> 전체 메뉴

menu = driver.find_element(By.XPATH,'//*[@id="gnb"]/div[2]/ul/li[3]/p/a')
all_menu = driver.find_element(By.XPATH,'//*[@id="gnb"]/div[2]/ul/li[3]/div/ul/li[1]/a')
```

# Action Chain으로 메뉴에 접근해보자

```python
# Action Chain이란?
    # selenium에서 제공하는 기능 중 하나, 마우스, 키보드의 액션을 생성할 수 있다.
    
from selenium.webdriver.common.action_chains import ActionChains
```

```python
# action 만들기 / 변수 생성
action = ActionChains(driver)

# 첫번째 관문, MENU로 이동
    # 사용자가 마우스를 menu버튼으로 이동하듯
    # 컴퓨터에게 Menu 버튼 위로 마우스 커서를 이동하라고 명령한 것
    # 우리는 이 행위를 코드로 작성한 것
action.move_to_element(menu)

# 두번째 관문,'전체메뉴 클릭'
action.click(all_menu)
time.sleep(1)

# 그리고, 해당 작업을 수행해줘
action.perform()
```

```python
# 더보기 버튼 수집
btn = driver.find_element(By.XPATH, '//*[@id="btn_more"]/span/a')
btn.click()
```

### 요소들 출력

```python
# 상품 제목을 뽑아보자
titles = driver.find_elements(By.CSS_SELECTOR,'h4.h.fz_03')
# 요소'들'이니까 title을 한번 추출해보자.
for i in titles:
    print(i.text)
```

```python
len(titles)
```

```python
# 가격을 추출해보자
prices = driver.find_elements(By.CSS_SELECTOR,'div.item-price')
for u in prices:
    print(u.text)
```

```python
# 엇 쌤, 처음에는 10개 항목만 추출이 됬는데, 다시 실행해보니 요소들이 다 뽑혀요

# 코드 문제는 아니다.

# 그 이유는
    # 클라이언트(브라우저)가 서버에서 이미 모든 데이터를 다 불러온 상태
    # 그렇기 때문에 브라우저를 종료한 후에도, 다시 접속해서
    # 더보기 버튼을 누르지 않아도 전체 목록을 열람할 수 있었다.
```

```python
# 더보기 버튼을 누르는데 지장이 있지 않을까?
    # 만약에 예외처리를 하지 않았다면, 문제발생 가능
    # 하지만, 우리는 곧 예외처리를 해줄거기 때문에,
        # 더보기 버튼이 없어도 크롤링이 끊기지 않고 진행되게 해줄거다
```

### 반복문을 통해서 데이터 크롤링 진행

```python

# 리스트 생성

titles_list = []
prices_list = []

try: #시도해봐 (아래 블럭의 코드를 실행해봐)
    while True : #(이 조건이 참일 때) (더보기 버튼이 있을 때)
        
        # 더보기 버튼 선언
        btn = driver.find.element(By.XPATH, '//*[@id="btn_more"]/span/a')
        # 더보기 버튼 클릭
        btn.click()
        # 지연시간을 2초 정도
        time.sleep(2)
        
        # 자 여기까지, 더보기 버튼이 계속 있다면, 클릭을 해주면서 계속 돌아갑니다.
except: # 예외의 경우
    
    # 이 실습에서 예외의 경우란, 더보기 버튼이 없는 경우
    # 예외처리가 안 되어 있을 경우, 더보기 버튼이 없을 시,
        # 동작은 되지만, 데이터가 잘 안뽑힐 수 있고,
        # 에러를 배출하면서 아예 크롤러가 종료가 될 수 있다.
    # 1. 우선 '클릭 완료''계속 진행합니다' 라는 문구를 출력
    print('더보기 클릭 완료, 계속 진행합니다.')
    
titles = driver.find_elements(By.CSS_SELECTOR,'h4.h.fz_03')
prices = driver.find_elements(By.CSS_SELECTOR,'div.item-price')

for i in tqdm(range(len(titles))) :
    titles_list.append(titles[i].text)
    prices_list.append(prices[i].text)
    
driver.quit()

print("크롤링 완료, 수고하셨습니다.")
```

## DataFrame에 담기

```python
len(titles_list),len(prices_list)
```

```python
# 딕셔너리로 우선 선언을 해준다.
    # 데이터 구조가 보존되므로,
        # Pandas DataFrame으로 변환할 떄 일관성이 유지된다
hansot ={"메뉴명":titles_list,'가격':prices_list}
```

```python
# DataFrame으로 변환을 해줘야함

hansot_df = pd.DataFrame(hansot)
hansot_df

# 혹시 중복되는 값이 있을까?
hansot_df.drop_duplicates()
hansot_df
```
