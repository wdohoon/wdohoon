# Q4-1

```python
from selenium import webdriver as wb
from bs4 import BeautifulSoup as bs
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
from tqdm import tqdm
import requests as req
import pandas as pd
```

```python
url = 'https://www.melon.com/chart/'
req.get(url)
head = {'User-Agent' : 'https://www.melon.com/chart/'}
res_melon = req.get('https://www.melon.com/chart/', headers = head)
soup = bs(res_melon.text,'lxml')


titles = soup.select('div.ellipsis.rank01 > span > a')
titles[0].text
artists = soup.select('div.ellipsis.rank02 > span')
titles_list = []
artists_list = []

for i in range(len(titles)):
    titles_list.append(titles[i].text)
    artists_list.append(artists[i].text)
    
    print(titles[i].text, artists[i].text)
melon_dic = {'title': titles_list,'artists' : artists_list}
melon_df = pd.DataFrame(melon_dic)
melon_df.to_csv('C:/Users/user07/webcrawling/melon_chart.csv',encoding = 'euc-kr')
```
