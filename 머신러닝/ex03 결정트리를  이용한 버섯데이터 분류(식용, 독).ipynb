### 모델링
- 데이터 분리 / 모델 선택 / 하이퍼 파라미터 튜닝 / 학습 / 평가 / 시각화

```python
# 데이터 분리
# 1. 문제와 정답 분리 (hint. 정답 컬럼 poisonous)
X = data.drop('poisonous', axis=1)
y = data.loc[:, 'poisonous']
```

```python
X.shape, y.shape
```

```python
# train_test_split을 통해서 데이터 분할
# 1.모듈 불러오기
from sklearn.model_selection import train_test_split
```

```python
# 2. 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                   test_size = 0.3,
                                                   random_state=14)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# 모델 불러오기
from sklearn.tree import DecisionTreeClassifier
```

```python
# 모델 객체 생성
tree_mush01 = DecisionTreeClassifier() # 기본 모델
```

```python
# 모델 학습
tree_mush01.fit(X_train, y_train)
```

- 머신러닝 모델은 숫자로 값을 입력받기를 기대한다.
- 문자값을 넣어주면 작동을 하지 못한다.
- 그래서 문자값을 숫자로 바꿔주는 인코딩이라는 작업이 필요하다.

**인코딩(encoding)**
- 범주형의 데이터를 수치형의 데이터로 바꿔주는 방법


**바이닝(binning)**
- 수치형의 데이터를 범주형으로 바꿔주는 방법

```python
# 1. Label인코딩
# 특성에 유니크한 값의 갯수에 따라 숫자를 1씩 증가시켜 mapping해주는 방법
# 주의사항) 회귀파트에서는 절대 사용 불가 - 숫자에 따른 순서나 중요도 인식가능성 있음
```

```python
# 예시)
X['cap-shape'].unique()
```

```python
# Before
X['cap-shape']
```

```python
# after
X['cap-shape'].map({'x' : 0, 'b' : 1, 's' : 2, 'f' : 3, 'k' : 4, 'c' : 5})
```

```python
# 레이블 인코딩을 손쉽게 하는 방법
from sklearn.preprocessing import LabelEncoder
```

```python
X['cap-shape'].value_counts()
```

```python
# 인코더 객체 생성
encoder = LabelEncoder()
# 인코더 학습
encoder.fit(X['cap-shape'])
# 데이터 변환 - transform : 학습된 인코더의 문자들을 숫자로 변환
label = encoder.transform(X['cap-shape'])
label
```

**원핫 인코딩**
- 모든 데이터를 0과 1로 변환 -> 레이블에 유니크값만큼 컬럼을 생성한 후 해당하는 데이터는 1 해당하지 않으면 0
- 컴퓨터는 2진법으로 모든 데이터들을 처리함. 일반적으로 레이블 인코딩 보다는 직관적으로 이해하기 쉬운 원핫 인코딩을 조금더 많이 사용한다.

```python
# 예시)
temp = pd.get_dummies(X[['cap-shape', 'cap-surface']])
temp
```

### 인코딩을 이용해서 에러를 해결해보자
- 원핫 인코딩 사용하기

```python
# 전체 문제데이터를 넣어주는 방법
X_one_hot = pd.get_dummies(X)
```

```python
# 만약 데이터가 분할되어있다면
# X_train = pd.get_dummies(X_train)
# X_test = pd.get_dummies(X_test)
```

### 데이터 전처리 및 변환시 주의점(엄청 중요!!!!!)
- 훈련셋과 평가셋을 전처리할때 같은 방법으로 전처리가 진행되어야 한다.(필수)
- 예시) 훈련용 문제데이터는 원핫인코딩 / 평가용 문제데이터에는 레이블 인코딩을 적용한 경우

```python
# 데이터 분할
X_train, X_test, y_train, y_test =train_test_split(X_one_hot, y,
                                                  test_size=0.3,
                                                  random_state=14)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# 모델 객체 생성
tree_mush01 = DecisionTreeClassifier()
```

```python
# 모델 학습
tree_mush01.fit(X_train, y_train)
```

```python
# 모델 평가
tree_mush01.score(X_test, y_test)
```

- 정확도가 100%???? 모델의 구조를 살펴보자(시각화)

```python
import graphviz
```

```python
# 모델 시각화
# 1. 모델 시각화 파일 만들기
from sklearn.tree import export_graphviz
```

```python
export_graphviz(tree_mush01, # 학습된 모델이 저장된 변수명
                out_file='tree01.dot', # 파일 명 및 저장 경로 설정
                class_names=['독', '식용'], # 클래스(정답, 레이블)이름 설정
                feature_names=X_one_hot.columns,
                impurity=True, # 지니불순도 표시여부
                filled=True # 노드에 색상을 채워줄 것인가?
               )
```

```python
import graphviz

# tree01.dot 불러와서 인코딩 시켜주고 f로 저장
with open('tree01.dot', encoding='UTF8') as f:
    dot_graph = f.read()
    
display(graphviz.Source(dot_graph))
```

```python
import graphviz

# tree02.dot 불러와서 인코딩 시켜주고 f로 저장
with open('tree02.dot', encoding='UTF8') as f:
    dot_graph = f.read()
    
display(graphviz.Source(dot_graph))
```
